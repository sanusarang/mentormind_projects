```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('heart_disease.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum()) # Checking for missing values
```

```python
duplicates = df.duplicated().sum() #removing duplicate rows
df.drop_duplicates(inplace=True)
print(f'Removed {duplicates} duplicate rows.')
```

```python
missing_values = df.isnull().sum()
print(f'Missing values in each column before handling:\n{missing_values}')
```

```python
df.fillna(df.median(), inplace=True) # Filling missing values (if any) with the median for numeric columns
```

```python
categorical_columns = ['HeartDiseaseorAttack', 'HighBP', 'HighChol', 'CholCheck', 'Smoker', 
                       'Stroke', 'Diabetes', 'PhysActivity', 'Fruits', 'Veggies', 
                       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex']

# Convert specified columns to categorical
for col in categorical_columns:
    df[col] = df[col].astype('category')

# Explicitly set the category order for 'Diabetes'
df['Diabetes'] = df['Diabetes'].astype('category').cat.reorder_categories([0, 1, 2], ordered=True)

# Convert 'Age', 'Education', and 'Income' columns to categorical data type
df['Age'] = df['Age'].astype('category')
df['Education'] = df['Education'].astype('category')
df['Income'] = df['Income'].astype('category')
```

```python
final_missing_values = df.isnull().sum()
final_duplicates = df.duplicated().sum()

print(f'Final missing values after handling:\n{final_missing_values}')
print(f'Final duplicate rows after cleaning: {final_duplicates}')
```

```python
df.info()
```

```python
# Set plot styles
sns.set(style="whitegrid")

# Plotting Histograms for numerical variables: BMI, MentHlth, PhysHlth
plt.figure(figsize=(14, 6))

# BMI distribution
plt.subplot(1, 3, 1)
sns.histplot(df['BMI'], kde=False, bins=30, color='skyblue')
plt.title('Distribution of BMI')

# MentHlth distribution
plt.subplot(1, 3, 2)
sns.histplot(df['MentHlth'], kde=False, bins=30, color='lightgreen')
plt.title('Distribution of Mental Health Days')

# PhysHlth distribution
plt.subplot(1, 3, 3)
sns.histplot(df['PhysHlth'], kde=False, bins=30, color='salmon')
plt.title('Distribution of Physical Health Days')

plt.tight_layout()
plt.show()

# Bar chart for Age distribution (since Age is categorical)
plt.figure(figsize=(10, 6))
sns.countplot(x='Age', data=df, palette='viridis')
plt.title('Distribution of Age Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting bar charts for the prevalence of health conditions: HighBP, HighChol, and others
health_conditions = ['HighBP', 'HighChol', 'Diabetes', 'Stroke', 'Smoker']

plt.figure(figsize=(14, 8))

for i, condition in enumerate(health_conditions, 1):
    plt.subplot(2, 3, i)
    if condition == 'Diabetes':
        sns.countplot(x=condition, data=df, palette='Set2')
        plt.title(f'Prevalence of {condition}')
        plt.xticks(ticks=[0, 1, 2], labels=['No Diabetes', 'Prediabetes', 'Diabetes'])
    else:
        sns.countplot(x=condition, data=df, palette='Set2')
        plt.title(f'Prevalence of {condition}')
        plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# Bar chart for the distribution of heart disease cases
plt.figure(figsize=(6, 4))
sns.countplot(x='HeartDiseaseorAttack', data=df, palette='Set1')
plt.title('Distribution of Heart Disease Cases')
plt.xticks(ticks=[0, 1], labels=['No Heart Disease', 'Heart Disease'])
plt.show()

# Pie chart for the distribution of heart disease cases
plt.figure(figsize=(6, 6))
df['HeartDiseaseorAttack'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], startangle=90)
plt.title('Proportion of Heart Disease Cases')
plt.ylabel('')  # Remove y-label for better presentation
plt.show()
```

```python
# Separate the variables into categorical and numeric
numeric_conditions = ['BMI']  # Keep only numeric variables
categorical_conditions = ['HighBP', 'HighChol']  # Use countplot for categorical variables

# Plot numeric variables with boxplot
plt.figure(figsize=(12, 6))
for i, condition in enumerate(numeric_conditions, 1):
    plt.subplot(1, len(numeric_conditions), i)
    sns.boxplot(x='HeartDiseaseorAttack', y=condition, data=df, palette='Set1')
    plt.title(f'Heart Disease vs {condition}')
    
plt.tight_layout()
plt.show()

# Plot categorical variables with countplot
plt.figure(figsize=(12, 6))
for i, condition in enumerate(categorical_conditions, 1):
    plt.subplot(1, len(categorical_conditions), i)
    sns.countplot(x=condition, hue='HeartDiseaseorAttack', data=df, palette='Set2')
    plt.title(f'Heart Disease vs {condition}')
    
plt.tight_layout()
plt.show()
```

```python
# Calculating the correlation matrix
corr_matrix = df[['Age', 'BMI', 'MentHlth', 'PhysHlth']].corr()

# Plotting the heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Continuous Variables')
plt.show()
```

```python
# Grouping data to analyze heart disease across demographic groups
demographic_groups = ['Age', 'Sex', 'Education']

plt.figure(figsize=(16, 12))
for i, group in enumerate(demographic_groups, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=group, hue='HeartDiseaseorAttack', data=df, palette='Set2')
    plt.title(f'Heart Disease Prevalence by {group}')
    plt.xticks(rotation=45)
    
plt.tight_layout()
plt.show()
```

```python
categorical_vars = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

# Create subplots for bar charts
plt.figure(figsize=(14, 8))
for i, var in enumerate(categorical_vars, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=var, data=df, palette='Set2')
    plt.title(f'Distribution of {var}')
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])

plt.tight_layout()
plt.show()

# Create pie charts for the same categorical variables
for var in categorical_vars:
    plt.figure(figsize=(6, 6))
    df[var].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
    plt.title(f'Distribution of {var}')
    plt.ylabel('')  # Remove the y-label for better aesthetics
    plt.show()
```

```python
# List of categorical variables to compare with Heart Disease
comparison_vars = ['Smoker', 'PhysActivity', 'Fruits', 'Veggies']

# Create grouped bar charts for Heart Disease vs other categorical variables
plt.figure(figsize=(14, 8))
for i, var in enumerate(comparison_vars, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=var, hue='HeartDiseaseorAttack', data=df, palette='Set1')
    plt.title(f'Heart Disease vs {var}')
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# Create stacked bar charts for Heart Disease vs categorical variables
for var in comparison_vars:
    # Create a crosstab to get the counts
    cross_tab = pd.crosstab(df[var], df['HeartDiseaseorAttack'])
    cross_tab.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(8, 6))
    
    plt.title(f'Heart Disease vs {var}')
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
    plt.ylabel('Count')
    plt.show()
```

```python
# List of continuous variables
continuous_vars = ['BMI', 'MentHlth', 'PhysHlth']

# Plotting histograms for continuous variables
plt.figure(figsize=(14, 8))
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[var], kde=True, color='skyblue')
    plt.title(f'Distribution of {var}')
    
plt.tight_layout()
plt.show()

# Plotting box plots for continuous variables
plt.figure(figsize=(14, 8))
for i, var in enumerate(continuous_vars, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='HeartDiseaseorAttack', y=var, data=df, palette='Set1')
    plt.title(f'Box Plot of {var} by Heart Disease')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = df[['BMI', 'MentHlth', 'PhysHlth', 'HighBP', 'HighChol', 'HeartDiseaseorAttack']].corr()

# Plotting a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Variable Correlations')
plt.show()
```

```python
# Scatter plots to investigate relationships between continuous variables and heart disease
plt.figure(figsize=(14, 8))
sns.scatterplot(x='BMI', y='MentHlth', hue='HeartDiseaseorAttack', data=df, palette='coolwarm')
plt.title('BMI vs Mental Health Days (MentHlth) with Heart Disease')
plt.show()

# Additional scatter plot example
plt.figure(figsize=(14, 8))
sns.scatterplot(x='BMI', y='PhysHlth', hue='HeartDiseaseorAttack', data=df, palette='coolwarm')
plt.title('BMI vs Physical Health Days (PhysHlth) with Heart Disease')
plt.show()
```

```python
# Ensure 'HeartDiseaseorAttack' is numeric
df['HeartDiseaseorAttack'] = pd.to_numeric(df['HeartDiseaseorAttack'], errors='coerce')

# Now calculate the mean heart disease rate across these age categories
age_heart_disease = df.groupby('Age')['HeartDiseaseorAttack'].mean().reset_index()

# Plot the line chart
plt.figure(figsize=(10, 6))
age_heart_disease.plot(x='Age', y='HeartDiseaseorAttack', kind='line', marker='o', color='blue')

# Customize the plot
plt.title('Heart Disease Prevalence Across Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Heart Disease Prevalence')
plt.xticks(rotation=45)
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Violin plot for continuous variables based on heart disease
plt.figure(figsize=(14, 8))
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df, palette='Set3')
plt.title('Violin Plot of BMI by Heart Disease')
plt.show()

# Violin plot for another variable
plt.figure(figsize=(14, 8))
sns.violinplot(x='HeartDiseaseorAttack', y='PhysHlth', data=df, palette='Set3')
plt.title('Violin Plot of Physical Health Days by Heart Disease')
plt.show()
```

```python
# Selecting relevant variables for pair plots
pairplot_vars = ['BMI', 'MentHlth', 'PhysHlth', 'HeartDiseaseorAttack']

# Plot pairwise relationships
sns.pairplot(df[pairplot_vars], hue='HeartDiseaseorAttack', palette='Set1')
plt.show()
```
